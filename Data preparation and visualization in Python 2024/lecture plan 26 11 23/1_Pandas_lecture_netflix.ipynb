## Aggregation 

Метод groub_by() позволяет группировать данные по каким-либо признакам и вычислять агрегированные значения

Например, средние, суммы, минимумы, максимумы и количество/ункальное количество 

Еще раз взглянем на датасет 

```python
data.head(2)
```

```python
data.index.name = "id"
data.reset_index(inplace = True)
data.head(1)
```

```python
pd.DataFrame(data.groupby(by = ["release_year", "type"])["id"].count()).reset_index()
```

```python
gdf = pd.DataFrame(data.groupby(by = "release_year")["id"].count()).reset_index()
```

```python
plt.plot(gdf["release_year"], gdf["id"])
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,3)
```

```python
plt.plot(gdf["release_year"], gdf["id"])
plt.show()
```

```python
ddf = data[data["genres"].str.contains("drama")].groupby(by = "release_year")["imdb_score"].mean()
ddf = pd.DataFrame(ddf).reset_index()
```

```python
plt.rcParams["figure.figsize"] = (15,5)
```

```python
plt.figure
plt.plot(ddf["release_year"], ddf["imdb_score"])
plt.title("Средняя оценка по карегории Драма")
plt.show()
```

```python
data["is_drama"] = data["genres"].str.contains("drama")
data["is_comedy"] = data["genres"].str.contains("comedy")
data["drama_and_comedy"] = data["is_drama"] & data["is_comedy"]
```

```python
ddf = data[ data["is_drama"] | data["is_comedy"] ].groupby(by = ["release_year", "is_drama", "drama_and_comedy"])["imdb_score"].mean()
ddf = pd.DataFrame(ddf).reset_index()
```

```python
ddf
```

```python
plt.figure
plt.plot(ddf["release_year"], ddf["imdb_score"])
plt.title("Средняя оценка по каретрии Драма")
plt.show()
```

```python
import seaborn as sns
```

```python
sns.relplot(data=ddf, x="release_year", y="imdb_score", hue="is_drama", kind="line")
plt.show()
```

```python
sns.relplot(data=ddf, x="release_year", y="imdb_score", hue="drama_and_comedy", kind="line")
plt.show()
```

```python
ddf
```

```python
plt.figure
plt.plot(ddf[ddf["is_drama"]]["release_year"], ddf[ddf["is_drama"]]["imdb_score"], label = "drama" )
plt.plot(ddf[~ddf["is_drama"]]["release_year"], ddf[~ddf["is_drama"]]["imdb_score"], label = "comedy" )
plt.plot(ddf[ddf["drama_and_comedy"]]["release_year"], ddf[ddf["drama_and_comedy"]]["imdb_score"], label = "drama and comedy" )
plt.legend()
plt.title("Average Imdb Score")
plt.show()
```
