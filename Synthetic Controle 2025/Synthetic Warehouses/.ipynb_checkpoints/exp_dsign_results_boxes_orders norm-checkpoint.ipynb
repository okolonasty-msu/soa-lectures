```python
d_boxes = d[d["type"] == "box"]
```

```python
def fill_nulls(X):
    X = (X.ffill()+X.bfill())/2
    X = X.ffill()
    X = X.bfill()
    return X
```

```python
df = d_boxes.pivot(columns = "route", index = "shipment_date", values = "orders")
df
```

for col in df.columns.values:
    if sum(df.tail(2)[col].isna()) > 0 or sum(df.head(1)[col].isna()) > 0:
        del df[col]

```python
df = fill_nulls(df)
#df
```

```python
iter_df = df.copy()
#iter_df
```

```python
corr = iter_df.corr()
#corr
```

```python
r = "Подольск 4 -> Краснодар (Тихорецкая)"
tmp = corr[[r]].sort_values(by = r, ascending=False)
#tmp = tmp.dropna()
tmp.fillna(0, inplace = True)
tmp
```

```python
controls = tmp[(tmp[r] >= 0.2) & (tmp[r] < 1)].index.values
controls
```

```python
print(r, controls, sep = "\n")

df_pre = iter_df[iter_df.index <= "2024-05-26"]
df_post = iter_df[iter_df.index > "2024-05-26"]

test_trainsit_id = r
#control__matching_transit_id = 218225

control_transit_ids = [i for i in controls if i != test_trainsit_id]
print(control_transit_ids)

test_pre, test_post = df_pre[[test_trainsit_id]], df_post[[test_trainsit_id]]
test_pre

control_pre, control_post = df_pre[control_transit_ids], df_post[control_transit_ids]
control_pre

y = test_pre
X = control_pre

# Настраиваем веса синтетического города с помощью линейной регрессии
regr = LinearRegression().fit(X, y)

synthetic_control_transit  = regr.predict(pd.concat([control_pre, control_post]))

iter_df["synthetic_control_transit"] = synthetic_control_transit

plot_model = iter_df[[test_trainsit_id, "synthetic_control_transit"]]
  

plt.figure

plt.plot(plot_model.index, plot_model["synthetic_control_transit"], label = "synthetic_control_transit" )
plt.plot(plot_model.index, plot_model[test_trainsit_id], label = f"test_transit {test_trainsit_id}")
#plt.plot(plot_model.index, plot_model[control__matching_transit_id], label = "control_transit_matching (218225)")
plt.title(f"Динамика заказов в день")
plt.axvspan(xmin = "2024-01-01", xmax = "2024-05-26", color='grey', alpha = 0.1, label = "tarin period") 
plt.axvspan(xmin = "2024-05-26", xmax = "2024-06-20", color='g', alpha = 0.1, label = "test period") 

plt.legend()


plt.show()
```

```python
r = "Подольск 4 -> Краснодар (Тихорецкая)"
iter_df = df.copy()
#iter_df = iter_df.dropna()
window = 7
for col in iter_df.columns.values:
    iter_df[col] = iter_df[col].rolling(window).mean()
corr = iter_df.corr()

iter_df = iter_df[window:]
iter_df = fill_nulls(iter_df)

tmp = corr[[r]].sort_values(by = r, ascending=False)
controls = tmp[(tmp[r] >= 0.1) & (tmp[r] < 1)].index.values
print(r, controls, sep = "\n")

df_pre = iter_df[iter_df.index <= "2024-05-26"]
df_post = iter_df[iter_df.index > "2024-05-26"]

test_trainsit_id = r
#control__matching_transit_id = 218225

control_transit_ids = [i for i in controls if i != test_trainsit_id]

test_pre, test_post = df_pre[[test_trainsit_id]], df_post[[test_trainsit_id]]
test_pre

control_pre, control_post = df_pre[control_transit_ids], df_post[control_transit_ids]
control_pre

y = test_pre
X = control_pre

# Настраиваем веса синтетического города с помощью линейной регрессии
regr = LinearRegression().fit(X, y)

synthetic_control_transit  = regr.predict(pd.concat([control_pre, control_post]))

iter_df["synthetic_control_transit"] = synthetic_control_transit

plot_model = iter_df[[test_trainsit_id, "synthetic_control_transit"]]
  

plt.figure

plt.plot(plot_model.index, plot_model["synthetic_control_transit"], label = "synthetic_control_transit" )
plt.plot(plot_model.index, plot_model[test_trainsit_id], label = f"test_transit {test_trainsit_id}")
#plt.plot(plot_model.index, plot_model[control__matching_transit_id], label = "control_transit_matching (218225)")
plt.title(f"Динамика заказов в день: сглаженные ряды (ma 7)")
plt.axvspan(xmin = "2024-01-01", xmax = "2024-05-26", color='grey', alpha = 0.1, label = "tarin period") 
plt.axvspan(xmin = "2024-05-26", xmax = "2024-06-20", color='g', alpha = 0.1, label = "test period") 

plt.legend()


plt.show()
```
